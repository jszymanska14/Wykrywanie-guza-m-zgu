```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Nazwy folderów z klasami
labels = {'no': 0, 'yes': 1}

# Rozmiar obrazu po przeskalowaniu
image_size = 224

# Funkcja do wczytywania danych z folderu
def load_data(folder_path, image_size):
    X = []
    y = []
    for label, label_index in labels.items():
        class_path = os.path.join(folder_path, label)
        print(f"Checking directory: {class_path}")
        if not os.path.exists(class_path):
            print(f"Directory does not exist: {class_path}")
            continue
        for filename in tqdm(os.listdir(class_path)):
            img_path = os.path.join(class_path, filename)
            print(f"Loading image: {img_path}")
            img = cv2.imread(img_path)
            if img is None:
                print(f"Failed to load image: {img_path}")
                continue
            img = cv2.resize(img, (image_size, image_size))
            img = np.expand_dims(img, axis=0)
            X.append(img)
            y.append(label_index)
    return np.array(X), np.array(y)


X_train, y_train = load_data('/https:/github.com/jszymanska14/Wykrywanie-guza-m-zgu.git/train', image_size)


X_test, y_test = load_data('/https:/github.com/jszymanska14/Wykrywanie-guza-m-zgu.git/test', image_size)


X_val, y_val = load_data('/https:/github.com/jszymanska14/Wykrywanie-guza-m-zgu.git/valid', image_size)


datasets = [('Trening', X_train, y_train), ('Testowy', X_test, y_test), ('Walidacyjny', X_val, y_val)]

# Zdefiniowanie słownika mapującego etykiety (0 i 1) na nazwy klas
labels = {0: 'no', 1: 'yes'}


for dataset_name, X, y in datasets:
   
    # Liczba obrazów dla każdej klasy
    counts = pd.Series(y).value_counts()
    print(f"Liczba obrazów dla każdej klasy w zbiorze {dataset_name}")
    print(counts)
    
    # Wykres słupkowy przedstawiający rozkład klas
    plt.figure(figsize=(6, 10))
    sns.countplot(x=y)
    plt.xlabel('Klasa')
    plt.ylabel('Liczba obrazów')
    plt.title(f'Rozkład klas w zbiorze {dataset_name}')
    plt.show()

    
    sizes = np.array([img.shape for img in X])
    print(f"\nStatystyki obrazów w zbiorze {dataset_name}")
    print(f"Średni rozmiar obrazów: {np.mean(sizes, axis=0)}")
    print(f"Maksymalny rozmiar obrazów: {np.max(sizes, axis=0)}")
    print(f"Minimalny rozmiar obrazów: {np.min(sizes, axis=0)}")

    # Wyświetlenie kilku przykładowych obrazów z różnych klas
    plt.figure(figsize=(10, 5))
    for i, label in enumerate([0, 1]): 
        
        index = np.where(y == label)[0][0]
       
        plt.subplot(1, 2, i + 1)
        plt.imshow(X[index][0]) 
                   #,cmap='gray')
        plt.title(f'Klasa: {labels[label]}')
        plt.axis('off')
    plt.suptitle(f'Przykładowe obrazy z klas w zbiorze {dataset_name}')
    plt.show()
```

```python
import glob
import numpy as np
import cv2
from torch.utils.data import Dataset, DataLoader
import torch
from torch import nn
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

class MRI(Dataset):
    def __init__(self, train_data_path=None, val_data_path=None):
        self.train_data = None
        self.val_data = None
        self.train_images = None
        self.val_images = None
        self.train_labels = None
        self.val_labels = None
        if train_data_path:
            self.load_train_data(train_data_path)
        if val_data_path:
            self.load_validation_data(val_data_path)
        
    def load_train_data(self, train_data_path):
        tumor = []
        healthy = []
        
        for f in glob.iglob(f'{train_data_path}/no/*.jpg'):
            img = cv2.imread(f)
            if img is not None:
                img = cv2.resize(img,(128,128)) 
                b, g, r = cv2.split(img)
                img = cv2.merge([r,g,b])
                img = np.transpose(img, (2, 0, 1))
                healthy.append(img)
            else:
                print("Error loading image:", f)

        for f in glob.iglob(f'{train_data_path}/yes/*.jpg'):
            img = cv2.imread(f)
            if img is not None:
                img = cv2.resize(img,(128,128)) 
                b, g, r = cv2.split(img)
                img = cv2.merge([r,g,b])
                img = np.transpose(img, (2, 0, 1))
                tumor.append(img)
            else:
                print("Error loading image:", f)

        tumor = np.array(tumor, dtype=np.float32)
        healthy = np.array(healthy, dtype=np.float32)
        
        tumor_label = np.ones(tumor.shape[0], dtype=np.float32)
        healthy_label = np.zeros(healthy.shape[0], dtype=np.float32)
        
        self.train_images = np.concatenate((tumor, healthy), axis=0)
        self.train_labels = np.concatenate((tumor_label, healthy_label))
        
        self.train_data = list(zip(self.train_images, self.train_labels))

    def load_validation_data(self, val_data_path):
        tumor = []
        healthy = []

        for f in glob.iglob(f'{val_data_path}/no/*.jpg'):
            img = cv2.imread(f)
            if img is not None:
                img = cv2.resize(img, (128, 128))
                b, g, r = cv2.split(img)
                img = cv2.merge([r, g, b])
                img = np.transpose(img, (2, 0, 1))
                healthy.append(img)
            else:
                print("Error loading image:", f)

        for f in glob.iglob(f'{val_data_path}/yes/*.jpg'):
            img = cv2.imread(f)
            if img is not None:
                img = cv2.resize(img, (128, 128))
                b, g, r = cv2.split(img)
                img = cv2.merge([r, g, b])
                img = np.transpose(img, (2, 0, 1))
                tumor.append(img)
            else:
                print("Error loading image:", f)

        tumor = np.array(tumor, dtype=np.float32)
        healthy = np.array(healthy, dtype=np.float32)

        tumor_label = np.ones(tumor.shape[0], dtype=np.float32)
        healthy_label = np.zeros(healthy.shape[0], dtype=np.float32)

        self.val_images = np.concatenate((tumor, healthy), axis=0)
        self.val_labels = np.concatenate((tumor_label, healthy_label))

        self.val_data = list(zip(self.val_images, self.val_labels))
        
    def __len__(self):
        if self.train_data is not None:
            return len(self.train_data)
        elif self.val_data is not None:
            return len(self.val_data)
        else:
            return 0
    
    def __getitem__(self, index):
        if self.train_data is not None:
            image, label = self.train_data[index]
        elif self.val_data is not None:
            image, label = self.val_data[index]
        else:
            raise ValueError("Ani train_data, ani val_data nie są dostępne.")
        return {'image': image, 'label': label}
        
    def len_val(self):
        return len(self.val_data)
    
    def get_val_item(self, index):
        image, label = self.val_data[index]
        return {'image': image, 'label': label}
    
    def normalize(self):
        if self.train_images is not None:
            self.train_images = self.train_images / 255.0
        if self.val_images is not None:
            self.val_images = self.val_images / 255.0
        
mri_dataset_train = MRI(train_data_path='/Users/juliaszymanska/PycharmProjects/projekt_sieci_neuronowe/train')
mri_dataset_val = MRI(val_data_path='/Users/juliaszymanska/PycharmProjects/projekt_sieci_neuronowe/valid')
mri_dataset_train.normalize()
mri_dataset_val.normalize()

train_dataloader = DataLoader(mri_dataset_train, batch_size=32, shuffle=True)
val_dataloader = DataLoader(mri_dataset_val, batch_size=32, shuffle=False)

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        
        self.cnn_model = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=6, kernel_size=5),
            nn.Tanh(),
            nn.AvgPool2d(kernel_size=2, stride=5),
            nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5),
            nn.Tanh(),
            nn.AvgPool2d(kernel_size=2, stride=5)
        )
        
        self.fc_model = nn.Sequential(
            nn.Linear(in_features=256, out_features=120),
            nn.Tanh(),
            nn.Dropout(0.5),
            nn.Linear(in_features=120, out_features=84),
            nn.Tanh(),
            nn.Dropout(0.5),
            nn.Linear(in_features=84, out_features=1)
        )
        
    def forward(self, x):
        x = self.cnn_model(x)
        x = x.view(x.size(0), -1)
        x = self.fc_model(x)
        x = torch.sigmoid(x)
        return x

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = CNN().to(device)
eta = 0.0001
optimizer = torch.optim.Adam(model.parameters(), lr=eta)
loss_function = nn.BCELoss()


early_stopping_patience = 10
min_val_loss = np.inf
patience_counter = 0
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=5, factor=0.5, verbose=True)

EPOCHS = 100
train_losses = []
val_losses = []
val_accuracies = []

for epoch in range(1, EPOCHS + 1):
    model.train()
    train_loss = []
    for batch in train_dataloader:
        images, labels = batch['image'].to(device), batch['label'].to(device)
        labels = labels.view(-1, 1)  
        optimizer.zero_grad()
        outputs = model(images)
        loss = loss_function(outputs, labels)
        loss.backward()
        optimizer.step()
        train_loss.append(loss.item())
    
    avg_train_loss = np.mean(train_loss)
    train_losses.append(avg_train_loss)

    model.eval()
    val_loss = []
    val_predictions = []
    val_labels = []

    with torch.no_grad():
        for batch in val_dataloader:
            images, labels = batch['image'].to(device), batch['label'].to(device)
            labels = labels.view(-1, 1)  
            outputs = model(images)
            loss = loss_function(outputs, labels)
            val_loss.append(loss.item())
            val_predictions.extend(outputs.cpu().numpy())
            val_labels.extend(labels.cpu().numpy())
    
    avg_val_loss = np.mean(val_loss)
    val_losses.append(avg_val_loss)
    
    val_predictions = [1 if pred >= 0.5 else 0 for pred in val_predictions]
    val_accuracy = accuracy_score(val_labels, val_predictions)
    val_accuracies.append(val_accuracy)
    
    print(f'Epoch {epoch}: Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}')

    # Wczesne zatrzymywanie
    scheduler.step(avg_val_loss)
    
    if avg_val_loss < min_val_loss:
        min_val_loss = avg_val_loss
        patience_counter = 0
        torch.save(model.state_dict(), 'best_model.pth')
    else:
        patience_counter += 1
    
    if patience_counter >= early_stopping_patience:
        print("Early stopping triggered.")
        break


model.load_state_dict(torch.load('best_model.pth'))


plt.style.use('dark_background')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

ax1.plot(train_losses, label='Train Loss')
ax1.plot(val_losses, label='Val Loss')
ax1.set_title('Train & Val Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()

ax2.plot(val_accuracies, label='Val Accuracy')
ax2.set_title('Val Accuracy')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.legend()

plt.show()


```
